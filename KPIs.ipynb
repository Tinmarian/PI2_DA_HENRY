# Desarrollo de KPIs

El objetivo es obtener un mínimo de 3 KPIs. Los siguientes son los que vamos a desarrollar.

        * Vamos a obtener la división entre el volumen y el market cap, utilizando datos de coingecko, para cada criptomoneda por separado. Este KPI nos va a informar sobre el performance del volumen de la criptomoneda con respecto al suministro total. O lo que es lo mismo, nos va a informar sobre la cantidad de criptomonedas tradeadas con respecto al suministro máximo. Si bien no encontré datos directos de dicho suministro, el market cap lo mide de manera indirecta. Lo vamos a tomar con base en la última fecha.

        * El siguiente será una conjunción de los KPIs individuales, esto es, el promedio total del conjunto de participaciones de cada cripto con respecto a su propio market cap.

        * Por último, vamos a obtener un promedio de capitalización de mercado de las criptomonedas trabajadas y se va a comparar con el market cap total del mercado cripto. También con respecto al dato de la última fecha.

```python
import os
import pandas as pd
```

```python
dfc = []
crl = []
for i in [i for i in os.listdir('CoinGecko/data') if '_' not in i]:
    path = f'CoinGecko/data/{i}'
    crl.append(i.replace('.csv',''))
    dfc.append(pd.read_csv(path))
    if crl[-1] == 'btc':
        dfc = [dfc.pop()] + dfc
        crl = [crl.pop()] + crl

j = 0
for i in dfc:
    i['Fecha'] = pd.to_datetime(i.Fecha)
    i = i[['Fecha','Market_Cap']]
    i.columns = ['Fecha',f'Market_Cap_{crl[j].upper()}']
    if j == 0:
        df_cgm = i
    else:
        df_cgm = df_cgm.merge(i,how='left',on='Fecha')
    j += 1
    
df_cgm.head(3)
```

```python
dfx = pd.DataFrame(df_cgm.iloc[-2,:]).reset_index()
for i in sorted(crl):
    dfx['index'] = dfx['index'].apply(lambda x: f'{i.upper()}' if i.upper() in x else x)
dfx = dfx.drop(0).reset_index(drop=True)
dfx
```

```python
dfc = []
crl = []
for i in [i for i in os.listdir('CoinGecko/data') if '_' not in i]:
    path = f'CoinGecko/data/{i}'
    crl.append(i.replace('.csv',''))
    dfc.append(pd.read_csv(path))
    if crl[-1] == 'btc':
        dfc = [dfc.pop()] + dfc
        crl = [crl.pop()] + crl

j = 0
for i in dfc:
    i['Fecha'] = pd.to_datetime(i.Fecha)
    i = i[['Fecha','Volumen']]
    i.columns = ['Fecha',f'USD_Volumen_{crl[j].upper()}']
    if j == 0:
        df_cgv = i
    else:
        df_cgv = df_cgv.merge(i,how='left',on='Fecha')
    j += 1
    
df_cgv.head(3)
```

```python
dfy = pd.DataFrame(df_cgv.iloc[-2,:]).reset_index()
for i in sorted(crl):
    dfy['index'] = dfy['index'].apply(lambda x: f'{i.upper()}' if i.upper() in x else x)
dfy = dfy.drop(0).reset_index(drop=True)
dfy
```
